# Label Encoding
- ### Converting labels into numerical forms

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
cancer_df = pd.read_csv('Dataset/cancer_dataset.csv')
cancer_df.head()
```

```python
cancer_df.diagnosis.value_counts()
```

```python
# Load the LabelEncoder object
label_encoder = LabelEncoder()
```

```python
labels = label_encoder.fit_transform(cancer_df.diagnosis)
```

```python
# appending the labels to the DataFrame
cancer_df['target'] = labels
```

```python
cancer_df.head()
```

```python
cancer_df['target'].value_counts()
```

```python
iris_df = pd.read_csv('iris_data.csv')
iris_df.head()
```

```python
iris_df.Species.value_counts()
```

```python
label_encoder2 = LabelEncoder()
labels2 = label_encoder2.fit_transform(iris_df.Species)
```
