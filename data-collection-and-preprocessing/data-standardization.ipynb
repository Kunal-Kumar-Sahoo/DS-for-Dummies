# Data standardization
#### The process of standardizing the data to a common format and common range

```python
import numpy as np
import pandas as pd

import sklearn.datasets
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
dataset = sklearn.datasets.load_breast_cancer()
print(dataset)
```

```python
breast_cancer_df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
print(breast_cancer_df.shape)
breast_cancer_df.head()
```

```python
X = breast_cancer_df
Y = dataset.target
```

```python
X # Design matrix
```

```python
Y # Output vector
```

Split the dataset into **Training** and  **Testing** data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=3)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

### Data standardization

```python
dataset.data.std()
```

The above value shows the data is not in the same range

standard score (z) of a sample x:
    z = (x - u) / s
    where: u -> Mean, s->standard deviation

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
X_train_standardized = scaler.transform(X_train)
```

```python
X_train_standardized
```

```python
X_test_standardized = scaler.transform(X_test)
```

```python
X_train_standardized.std(), X_test_standardized.std()
```

Above values being close to 1 shows that the data is more standardized
